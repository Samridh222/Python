```python
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

Filter for US data analyst roles..

```python
df_DA_German = df[(df['job_country'] == 'Germany') & (df['job_title_short'] == 'Data Analyst')]
```

```python
df_plot = df_DA_German['job_location'].value_counts().head(10).to_frame()
import seaborn as sns
sns.barplot(data=df_plot, x='count', y='job_location', hue='count', palette='dark:b_r', legend=False)
sns.set_theme(style='ticks')
sns.despine()
plt.title('Job Location for Data Analyst in Germany')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```

Count of Companis for Data Analyst in Germany

```python
df_plot = df_DA_German['company_name'].value_counts().head(10).to_frame()
import seaborn as sns
sns.barplot(data=df_plot, x='count', y='company_name', hue='count', palette='dark:b_r', legend=False)
sns.set_theme(style='ticks')
sns.despine()
plt.title('Name of Companies for Data Analyst in Germany')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```
